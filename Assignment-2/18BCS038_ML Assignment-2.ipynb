```python
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
N=3
def f(x):
    y=2.21*x-2.32;
    return y
f2 = np.vectorize(f)
x=[1,2,3]
y=[1.2,1.9,3.2]
plt.title('Maximum Likelihood linear function')
plt.xlabel('data points')
plt.ylabel('fitted target points')
plt.plot(x,f2(x))
plt.plot(x,f2(x),'ro')
plt.plot(x,y,'bo')
plt.show()
```
